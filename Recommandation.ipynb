```python
import pandas as pd
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
```

```python
movies_df=pd.read_csv('movies.csv')
ratings_df=pd.read_csv('ratings.csv')
```

```python
print(movies_df.head())
```

# we want to move year in "title" column to another column
so we have to use regex(regular expression)

```python
movies_df['year']=movies_df.title.str.extract('(\(\d\d\d\d\))', expand=False)
movies_df['year']= movies_df.year.str.extract('(\d\d\d\d)', expand=False)
movies_df['title']=movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df['title']=movies_df['title'].apply(lambda x: x.strip())
```

```python
print(movies_df.head())
```

#split the values in the genres column into a list of genres

```python
movies_df['genres']=movies_df.genres.str.split('|')
print(movies_df.head())
```

# if the film has this genres shows 1 if doesnot have shows 0

```python
movies_df= movies_df.copy()
for index, row in movies_df.iterrows():
  for genres in row['genres']:
    movies_df.at[index, genres]=1
    movies_df=movies_df.fillna(0)
```

#we want to merge inputId and inputMovies

```python
userInput = [{'title': 'Breakfast Club, The', 'rating':5},
 {'title': 'Toy Story', 'rating':3.5},
             {'title': 'Jumanji', 'rating':2},
              {'title': "Pulp Fiction", 'rating':5},
              {'title':'Akira', 'rating':4.5}]
```

```python
inputId = movies_df[movies_df['title'].isin(inputmovies['title'].tolist())]
inputMovies = pd.merge(inputId, inputMovies)
inputMovies=inputMovies.drop('genres', 1)
inputMovies
```
