```python
import pandas as pd
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
```

```python
movies_df=pd.read_csv('movies.csv')
ratings_df=pd.read_csv('ratings.csv')
```

```python
print(movies_df.head())
```

# we want to move year in "title" column to another column
so we have to use regex(regular expression)

```python
movies_df['year']=movies_df.title.str.extract('(\(\d\d\d\d\))', expand=False)
movies_df['year']= movies_df.year.str.extract('(\d\d\d\d)', expand=False)
movies_df['title']=movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df['title']=movies_df['title'].apply(lambda x: x.strip())
```

```python
print(movies_df.head())
```

#split the values in the genres column into a list of genres

```python
movies_df['genres']=movies_df.genres.str.split('|')
print(movies_df.head())
```

# if the film has this genres shows 1 if doesnot have shows 0

```python
movies_df= movies_df.copy()
for index, row in movies_df.iterrows():
  for genres in row['genres']:
    movies_df.at[index, genres]=1
    movies_df=movies_df.fillna(0)
```

```python
movies_df.head()
```

```python
ratings_df.head()
```

#we want to merge inputId and inputMovies

```python
userInput = [{'title': 'Breakfast Club, The', 'rating':5},
 {'title': 'Toy Story', 'rating':3.5},
             {'title': 'Jumanji', 'rating':2},
              {'title': "Pulp Fiction", 'rating':5},
              {'title':'Akira', 'rating':4.5}]
```

```python
inputmovies=pd.DataFrame(userInput)
inputmovies
```

#we want to add movie Id to input user

```python
inputId = movies_df[movies_df['title'].isin(inputmovies['title'].tolist())]
inputmovies = pd.merge(inputId, inputmovies)
inputmovies=inputmovies.drop('genres', 1).drop('year', 1)
inputmovies
```

# clean data by resetting data and dropping the movieId, title, genres and year columns.

```python
usermovies= inputmovies.reset_index(drop=True)
userGenreTable= usermovies.drop('movieId', 1). drop('title', 1)
userGenreTable
```

# make user profile

```python
inputmovies['rating']
```

```python
userProfile = userGenreTable.transpose().dot(inputmovies['rating'])
userProfile
```

#get the ganres of every movie in our dataframe

```python
genreTable=  movies_df.set_index( movies_df['movieId'])
genreTable= genreTable.reset_index(drop=True)
genreTable = genreTable.drop('movieId', 1). drop('title', 1). drop('genres', 1). drop('year', 1)
genreTable
```

# make recommendation table

```python
recommendationTable= ((genreTable*userProfile). sum(axis=1))/(userProfile.sum())
recommendationTable.head()
```

#sort our recommendation indescending order:

```python
recommendationTable= recommendationTable.sort_values(ascending=False)
recommendationTable.head()
```

#the best name of film in recommendation table

```python
movies_df.loc[movies_df['movieId'].isin(recommendationTable.head(20). keys())]
```
